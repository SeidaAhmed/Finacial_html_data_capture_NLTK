```python
import google.generativeai as genai
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from markdownify import markdownify
import json
import os
```

```python
def extract_documents(url):
    headers = {
        "User-Agent": "UbineerCorp info@ubineer.com",
        "Accept-Encoding": "gzip, deflate",
        "Host": "www.sec.gov"
    }
    response = requests.get(url, headers=headers)
    text = response.text

    # Extract all <DOCUMENT> ... </DOCUMENT> blocks
    documents = re.findall(r"<DOCUMENT>(.*?)</DOCUMENT>", text, flags=re.DOTALL | re.IGNORECASE)
    return documents

def parse_document(doc_text):
    # Parse with BeautifulSoup
    soup = BeautifulSoup(doc_text, "html.parser")

    # Full HTML of the document
    full_html = str(soup)

    # Extract all tables as HTML strings
    tables_html = [str(table) for table in soup.find_all("table")]

    # Extract data from tables (example: extracting all text within <td> tags)
    tables_data = []
    for table in soup.find_all("table"):
        table_data = []
        for row in table.find_all("tr"):
            row_data = [cell.get_text(strip=True) for cell in row.find_all(["td", "th"])]
            table_data.append(row_data)
        tables_data.append(table_data)

    return full_html, tables_html, tables_data

def parse_markdown_table(md_table):
    """
    Parse a markdown table string into a list of rows (list of lists).
    """
    rows = []
    for line in md_table.splitlines():
        # Skip separator lines (---|---)
        if re.match(r'^\s*\|?\s*[-:]+\s*(\|[-:]+\s*)*\|?\s*$', line):
            continue
        if "|" in line:
            cells = [cell.strip() for cell in line.strip("|").split("|")]
            rows.append(cells)
    return rows

def discover_table_index_list(tables_data, required_features, threshold=0.5):
    """
    Locate index of the first table where most required features appear
    in the first column.
    """
    for idx, table in enumerate(tables_data):
        if not table:  # skip empty tables
            continue

        # Extract the first column of the table
        first_col = [row[0] for row in table if row]

        # Count how many required features appear in the first column
        matches = sum(1 for feature in required_features if feature in first_col)

        if matches / len(required_features) >= threshold:
            return idx
    return -1

required_features1 = ["Adjusted cost of revenue", "Gross bookings","B2C","B2B","Revenue","other","%Revenue","%B2C revenue","%B2B revenue","Selling and marketing-direct","Other segment items"]
required_features=['Gross Booking','B2C','B2B',
  'Revenue',
  'Revenue_B2C',
  'Revenue_B2B',
  'Other(1)',
  'Revenue margin(2)' ,
  'Adjusted cost of revenue(3)',
  '% Revenue',
  'Adjusted cost of revenue(3)_B2C',
  '% B2C revenue',
  'Adjusted cost of revenue(3)_B2B',
  '% B2B revenue',
  'Other(1)_revenue',
  'Selling and marketing - direct',
  '% Gross bookings',
  'Selling and marketing - direct_B2C',
  '% B2C gross bookings',
  'Selling and marketing - direct_B2B',
  'Selling and marketing - direct_Other(1)',
  'Other segment items(4)',
  'Other segment items(4)_B2C',
  'Other segment items(4)_B2B',
  'Other segment items(4)_Other(1)',
  'Adjusted EBITDA(3)',
  '% Margin',
  'Adjusted EBITDA(3)_B2C',
  'Adjusted EBITDA(3)_B2B',
  'Adjusted EBITDA(3)_Other(1)']
url = "https://www.sec.gov/Archives/edgar/data/1324424/000132442425000039/earningsrelease-q22025.htm"
docs = extract_documents(url)

for i, doc in enumerate(docs, 1):
    print(f"--- Document {i} ---")
    full_html, tables_html, tables_data = parse_document(doc)
    print("Full HTML preview:", full_html[:500])
    print("Number of tables:", len(tables_html))

    table_index = discover_table_index_list(tables_data, required_features)
    if table_index != -1:
        print(f"Required table found at index {table_index}")
        print("Table preview:", tables_data[table_index][:5])
    else:
        print("Required table not found")
```
